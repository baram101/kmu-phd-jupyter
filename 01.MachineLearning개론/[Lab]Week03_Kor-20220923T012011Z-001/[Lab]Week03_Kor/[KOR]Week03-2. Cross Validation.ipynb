# Simple KFold

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=3, shuffle=True)
```

```python
X_columns = [col for col in df.columns if col != 'Survived']
y_columns = ['Survived']
```

```python
trn_sets = []
tst_sets = []

# kf.split: return the index of train / test dataset
for i, (trn_idxs, tst_idxs) in enumerate(kf.split(df[X_columns], df[y_columns])):
    print(f'Train set ({i+1}) indexes: ')
    print(trn_idxs)
    print(f'Test set ({i+1}) indexes: ')
    print(tst_idxs)

    trn_sets.append(df.loc[trn_idxs])
    tst_sets.append(df.loc[tst_idxs])
    
```

```python
tst_sets[2]
```

- <span style = 'font-size:1.4em;line-height:1.5em'><b>분할된 데이터셋에는 클래스(레이블)의 비율이 다릅니다.</b></span>

```python
print('Class ratio for every splitted dataset')

for i in range(3):
    df_trn = trn_sets[i]
    df_tst = tst_sets[i] 
    
    df_trn_class1 = df_trn[df_trn['Survived'] == 1]
    df_tst_class1 = df_tst[df_tst['Survived'] == 1]
    
    print(f"Split {i+1}")
    print(f"# of Train dataset: {df_trn.shape[0]}")
    print(f"# of Test dataset: {df_tst.shape[0]}")
    print(f"# of Survived=1 in Train dataset: {df_trn_class1.shape[0]}")
    print(f"# of Survived=1 in Test dataset: {df_tst_class1.shape[0]}")
    print(f"The ratio of Survied=1 in Train dataset: {df_trn_class1.shape[0] / df_trn.shape[0]}")
    print(f"The ratio of Survied=1 in Test dataset: {df_tst_class1.shape[0] / df_tst.shape[0]}")
```

# Stratified KFold

- <span style = 'font-size:1.3em;line-height:1.5em'><b>분할된 데이터셋에서 동일한 레이블 비율을 가질 수 있도록 데이터를 분할하는 방법입니다.</b></span>

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kf_stratified = StratifiedKFold(n_splits=3, shuffle=True)
```

```python
X_columns = [col for col in df.columns if col != 'Survived']
y_columns = ['Survived']
```

```python
trn_sets = []
tst_sets = []

# kf.split: return the index of train / test dataset
for i, (trn_idxs, tst_idxs) in enumerate(kf_stratified.split(df[X_columns], df[y_columns])):
    print(f'Train set ({i+1}) indexes: ')
    print(trn_idxs)
    print(f'Test set ({i+1}) indexes: ')
    print(tst_idxs)

    trn_sets.append(df.loc[trn_idxs])
    tst_sets.append(df.loc[tst_idxs])
    
```

```python
print('Class ratio for every splitted dataset')

for i in range(3):
    df_trn = trn_sets[i]
    df_tst = tst_sets[i] 
    
    df_trn_class1 = df_trn[df_trn['Survived'] == 1]
    df_tst_class1 = df_tst[df_tst['Survived'] == 1]
    
    print(f"Split {i+1}")
    print(f"# of Train dataset: {df_trn.shape[0]}")
    print(f"# of Test dataset: {df_tst.shape[0]}")
    print(f"# of Survived=1 in Train dataset: {df_trn_class1.shape[0]}")
    print(f"# of Survived=1 in Test dataset: {df_tst_class1.shape[0]}")
    print(f"The ratio of Survied=1 in Train dataset: {df_trn_class1.shape[0] / df_trn.shape[0]}")
    print(f"The ratio of Survied=1 in Test dataset: {df_tst_class1.shape[0] / df_tst.shape[0]}")
```

# Cross Validation

- <span style = 'font-size:1.3em;line-height:1.5em'>아직 모델에 대해서는 배우지 않았기 때문에, survival 확률에 대해서는 랜덤으로 생성하여 CV를 진행해보겠습니다.</span>

```python
from sklearn.metrics import roc_auc_score
```

```python
kf = StratifiedKFold(n_splits=5, shuffle=True)
```

```python
X_columns = [col for col in df.columns if col != 'Survived']
y_columns = ['Survived']
```

```python
results = []
for i, (trn_idxs, tst_idxs) in enumerate(kf.split(df[X_columns], df[y_columns])):
    print(f"CV ({i}): ", end='')
    tst_data = df.loc[tst_idxs]
    tst_real_y = tst_data[y_columns].values
    tst_pred_y = np.random.uniform(low = 0.3, high=0.7, size=tst_real_y.shape) # Randomly generate prediction results
    print(tst_real_y.shape, tst_pred_y.shape)
    auroc = roc_auc_score(tst_real_y, tst_pred_y)
    results.append(auroc)
```

```python
np.mean(results)
```

### 만약 sklearn의 모델을 사용하였다면, cross_val_score라는 함수를 사용하여 CV한 결과를 바로 얻을 수 있습니다.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_wine
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd

# 데이터 불러오기
np_data = load_wine()
wine_data = np_data.data
wine_label = np_data.target

data = pd.DataFrame(data = wine_data, columns = np_data.feature_names,)
data['label'] = np_data.target

features = data.iloc[:,:-1]
label = pd.DataFrame(data['label'])

# 모델 선언
model = DecisionTreeClassifier()

score = cross_val_score(model, features, label, scoring = 'accuracy', cv=5)
print(f"단일 accuracy : {score}")
print(f"5번의 fold 후 평균 accuracy: {np.mean(score)}")
```
