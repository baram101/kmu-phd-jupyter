```python
df = pd.read_csv('data/train.csv')
```

## 1. Passenger Class(Pclass)별로 생존 비율을 구해보세요.
- <span style = 'font-size:1.2em;line-height:1.5em'>Hint: 각 passenger class별로 총 탑승객의 수를 구하고, 생존한 탑승객의 수를 구한뒤, (생존한 탐승객 수) / (총 탑승객 수)를 passenger class별로 계산하면 됩니다. </span>

```python
def get_survived_ratio(pclass):
    pclass_cnt = df[(df['Pclass']==pclass)]['PassengerId'].count()
    pclass_srv_cnt = df[(df['Pclass']==pclass) & (df['Survived']==1)]['PassengerId'].count()
    pclass_srv_ratio = pclass_srv_cnt/pclass_cnt
    return pclass_srv_ratio

for a in df['Pclass'].unique():
    print("Pclass {} 생존비율 : {}".format(a,get_survived_ratio(a)))
```

## 2. 나이가 10살 미만(Age<10)인 사람과 10살 이상 (Age>=10)인 사람의 생존 비율을 구해보세요.

```python
young_cnt = df[(df['Age']<10)]['PassengerId'].count()
young_srv_cnt = df[(df['Age']<10) & (df['Survived']==1)]['PassengerId'].count()
young_srv_ratio = young_srv_cnt/young_cnt
print(young_srv_ratio)
```

## 3. 성별이 남성(Sex='Male')인 사람과 여성(Sex='Female')인 사람의 생존 비율을 구해보세요.

```python
def get_sex_survived_ratio(sex):
    sex_cnt = df[(df['Sex']==sex)]['PassengerId'].count()
    sex_srv_cnt = df[(df['Sex']==sex) & (df['Survived']==1)]['PassengerId'].count()
    sex_srv_ratio = sex_srv_cnt/sex_cnt
    return sex_srv_ratio

for a in df['Sex'].unique():
    print("Sex {} 생존비율 : {}".format(a,get_sex_survived_ratio(a)))
```
