````python
group_cnt = group.agg('count')
````

````python
group_cnt
````

### 특정 column에 대해서 통계량을 볼수도 있다.

````python
group_fare_avg
````

### Column별로 다른 요약 통계량을 볼 수도 있다.

# 3. apply lambda

- <span style = 'font-size:1.2em;line-height:1.5em'>apply lambda 식을 활용하여 dataframe에 있는 column에 함수를 적용하여 새로운 column을 생성할 수 있음</span>
    - <span style = 'font-size:1.1em;line-height:1.3em'>여기서 lambda는 Python에서 배우는 익명함수 lambda와 같은 같음</span>

- <span style = 'font-size:1.2em;line-height:1.5em'>일반적으로 파이썬에서 lambda함수는 map 함수와 함께 다음과 같이 사용
</span>

```python
a = [1,2,3]
a_square = map(lambda x: x**2, a)
list(a_square)
```

- <span style = 'font-size:1.2em;line-height:1.5em'>DataFrame에서 lambda식은 위의 lambda식을 거의 그대로 적용한 것
</span>
- <span style = 'font-size:1.2em;line-height:1.5em'>다만, 사용되는 형태가 조금 다른데, 아래와 같이 사용
</span>

```python
df['new_col'] = df['col'].apply(lambda x: my_func(x))
```

<span style = 'font-size:1.3em;line-height:1.5em'><b>사용 예시</b></span>

````python
df.head()
````

### 이름의 길이를 나타내는 column을 생성하고 싶다고 하자.

````python
df['Name_len'] = df['Name'].apply(lambda x: len(x))
````

````python
df.head()
````

### 이름의 길이가 30자 이상이면 'long name', 그렇지 않으면 'short name'이라는 값이 들어있는 column을 만들자.

````python
df['Name_len'] = df['Name'].apply(lambda x: 'long name' if len(x) >=30 else 'short name')
````

````python
df.head()
````

# 연습문제

````python
df = pd.read_csv('data/train.csv')
````

<span style = 'font-size:1.3em;line-height:1.5em'><b>1. age를 입력받고, age<10이면 group1, 10<=age<30이면 group2, 30<=age<50이면 group3, age>=50이면 group4를 return하는 함수를 작성하세요.</b></span>

````python
def get_age_group(age):
    if age < 10:
        ans = 'group1'
    elif 10<=age<30:
        ans = 'group2'
    elif 30<=age<50:
        ans = 'group3'
    else:
        ans = 'group4'
    return ans
````

````python
get_age_group(10.5)
````

<span style = 'font-size:1.3em;line-height:1.5em'><b>2. 위에서 작성한 함수를 활용하여 age를 그룹화한 column인 age_group이라는 column을 만드세요.</b></span>

````python
df['age_group'] = df['Age'].apply(lambda x: get_age_group(x))
````

````python
df.head()
````

<span style = 'font-size:1.3em;line-height:1.5em'><b>3. age_group별로 각 그룹에 해당하는 인원수와 average Fare를 구하세요. </b></span>

````python
group = df.groupby(by='age_group')
````

````python
agg_format = {'PassengerId': 'count', 'Fare': 'mean'}
group_info = group[['PassengerId', 'Fare']].agg(agg_format)
group_info
````

<span style = 'font-size:1.3em;line-height:1.5em'><b>1. apply lambda는 2개의 column을 입력받아서 사용할 수도 있습니다. 해당 방법을 구글링해서 아래 markdown cell에 작성하세요.</b></span>

df['new_col'] = df.apply(lambda x: my_func(x.col1, x.col2),axis=1)

<span style = 'font-size:1.3em;line-height:1.5em'><b>2. Age와 Pclass를 입력받고, 다음과 같은 조건을 만족하는 함수를 작성하세요.</b></span>

- <span style = 'font-size:1.1em;line-height:1.3em'>group1: Age<=10, Pclass=1</span>
- <span style = 'font-size:1.1em;line-height:1.3em'>group2: Age> 10, Pclass=1</span>
- <span style = 'font-size:1.1em;line-height:1.3em'>group3: Age<=10, Pclass=2</span>
- <span style = 'font-size:1.1em;line-height:1.3em'>group4: Age> 10, Pclass=2</span>
- <span style = 'font-size:1.1em;line-height:1.3em'>group5: Age<=10, Pclass=3</span>
- <span style = 'font-size:1.1em;line-height:1.3em'>group6: Age> 10, Pclass=3</span>    


````python
def get_age_pclass_group(age, pclass):
    if pclass==1:
        rtn_val = 'group1' if age<=10 else 'group2' if age>10 else 'no_group'
    elif pclass==2:
        rtn_val = 'group3' if age<=10 else 'group4' if age>10 else 'no_group'
    elif pclass==3:
        rtn_val = 'group5' if age<=10 else 'group6' if age>10 else 'no_group'
    else:
        rtn_val = 'no_group'
    return rtn_val
````

<span style = 'font-size:1.3em;line-height:1.5em'><b>3. 위에서 작성한 함수를 활용하여 age_pclass조합을 그룹화한 column인 age_pclass_group이라는 column을 만드세요.</b></span>


````python
df['age_pclass_group'] = df.apply(lambda x: get_age_pclass_group(x.Age, x.Pclass), axis=1)
df
````

<span style = 'font-size:1.3em;line-height:1.5em'><b>4. age_pclass_group별로 각 그룹에 해당하는 인원수와 average Fare를 구하세요. </b></span>


````python
pclass_group = df.groupby(by='age_pclass_group')
pclass_group
````

````python
agg_format = {'PassengerId': 'count', 'Fare': 'mean'}
group_info = pclass_group[['PassengerId', 'Fare']].agg(agg_format)
group_info
````
